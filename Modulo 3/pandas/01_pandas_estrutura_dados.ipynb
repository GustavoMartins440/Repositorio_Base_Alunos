# 1- Pandas e Estrutura de Dados - Series e DataFrame

Estrutura de dados fundamentais: Series e Dataframe. No Pandas, uma Serie é uma estrutura unidimensional , parecida com uma coluna de uma tabela: contém uma sequência de valores com um índice associado. Já um DataFrame é uma estrutura bidimensional, tabela ou matriz, formada por várias colunas (que são internamente, series alinhadas pelo mesmo índice).



```python
# Primeiro fazemos a instalação das bibliotecas e depois a importação das mesmas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criar uma Serie (uma única coluna)
s = pd.Series([10,20,30,40,50], name = 'Valores')
```

```python
type(s)
```

```python
print('\n____________________Exemplo de Series______________________\n')
print(s)
print('\n')
# apresentará os dados da lista no formato de uma serie com indices(indexados)
```

# 3. Seleção e Indexação com Pandas

Existem várias maneiras de selecionar dados de um DataFrame. As mais comuns são loc (baseada em rótulo) e iloc (baseada em posição/índice numéricos)

```python
# Colocamos os dados no pandas para manipular os dados (limpeza, etc)
# Visualizar as primeiras linhas
df_dsa.head(6) # o padrão é retornar 5 linhas
```

```python
# Selecionar uma única coluna (retorna uma Serie)
nomes = df_dsa['Nome']
print('\n_____________________Selecionar a coluna Nome_______________________\n')
print(nomes)
print('\n')
```

```python
type(nomes)
```

```python
# Selecionar múltiplas colunas (retorna um DataFrame)
info_pessoal = df_dsa[['Nome','Idade']]
print('\n____________________Selecionar as colunas Nome e Idade________________________\n')
print(info_pessoal)
print('\n')
```

```python
type(info_pessoal)
```

```python
# Visualizar .loc para selecionar pela linha (rótulo/indice) e a coluna(Nome)
nome = df_dsa.loc[1,'Nome']
print('\n__________Selecionar com .loc[1,"Nome"]______________________\n', (nome),'\n')
```

```python
# Usando .iloc para selecionar pela posição da linha (linha 2) e da coluna (coluna 3)
salario_fabiana = df_dsa.iloc[2,3]
print('\n__________Selecionar com .iloc[2,3]______________________\n','Salário R$',(salario_fabiana),'\n')
```

```python
# Selecionar um intervalo de linhas
primeiras_tres_linhas = df_dsa.loc[0:2] # a função .loc inclui a última linha
print('\n__________Selecionar as 3 primeiras linhas com .loc______________________\n')
print(primeiras_tres_linhas)
```

# 6. Operações e Tranformações

Você pode facilmente criar novas colunas a partir de outras ou aplicar funções para modificar dados. Fazemos isso com frequência em tarefas de engenharia de atributos em projetos de Ciências de Dados

```python
df_dsa.head()
```

```python
df_dsa['Salário Anual' ]= df_dsa['Salario' ]* 12 # isso que fizemos chama-se engenharia de atributos
print('\n______DataFrame com nova coluna "Salário Anual"_______\n')
print(df_dsa)
print('\n')
```

```python
df_dsa['Bonus' ]=df_dsa['Salario' ]. apply(lambda x: x * 0.10 if x > 7000 else x * 0.05)
print('\n__________DataFrame com nova coluna "Bônus"_______\n')
print(df_dsa)
print('\n')
```

Criar uma coluna de faixa etária

```python
condicoes = [
    df_dsa['Idade']<18,
    (df_dsa['Idade']>=18)&(df_dsa['Idade']<=30),
    (df_dsa['Idade']>30)&(df_dsa['Idade']<=60),
    df_dsa['Idade']>60
]
```

```python
print(condicoes)
```

```python
faixas = ['Menor de idade', 'Jovem', 'Adulto', 'Idoso']
```

```python
df_dsa['Faixa Etária' ]= np.select(condicoes, faixas, default='Idade Nao Informada')
```

# 8. Manipilação  de tipos de Dados com Pandas

Garantir que casa coluna tenha o tipo de dado correto é fundamental para qualquer tipo de análise

```python
print('\n___________Tipos de Dados Antes da Conversão_____________\n')
print(df_dsa.dtypes)
print('\n')
```

```python
df_dsa = df_dsa.dropna(subset=['Idade'])
df_dsa['Idade']=df_dsa['Idade'].astype(int)
```

```python
print('\n___________Tipos de Dados Antes da Conversão_____________\n')
print(df_dsa.dtypes)
print('\n')
```

```python
df_dsa['Nome']=df_dsa['Nome'].astype('string')
df_dsa['Cidade']=df_dsa['Cidade'].astype('string')
df_dsa['Faixa Etária']=df_dsa['Faixa Etária'].astype('string')
```

```python
print('\n___________Tipos de Dados Antes da Conversão_____________\n')
print(df_dsa.dtypes)
print('\n')
```

    Diferença entre object e string no pandas:

- object é um tipo genérico que armazena qualquer coisa em Python( strings, números, listas etc).
- string(desde o Pandas 1.0) é um tipo de dado nativo para texto, pensado para operações mais seguras e consistentes com strings.
Com strings voce ganha melhor integracao com metodos de texto(.str) e tratamento explícito de valores ausentes ( em de NaN
misturado com objetos). Em grandes DataFrames, o tipo string pode ser um pouco mais eficiente e evita confusões quando cluas
deveriam ser só texto. Mas não é obrigatório converter, mas usar string ajuda a ter consistência, evita misturar tipos diferentes na
mesma coluna e facilita operações textuais. Para análises e limpeza de dados de texto, geralmente vale a pena.

# 9. Criar tabelas Dinâmica (Pivolt Tables) com Pandas

Tabelas dinâmicas são ótimas para resumir dados de forma semelhante a uma planilha

```python
dados_vendas = {
    'Data': pd.to_datetime(['2025-09-01','2025-09-01','2025-09-02','2025-09-01', '2025-09-01']),
    'Região':['Norte','Sul','Norte', 'Sul','Norte'],
    'Vendedor':['Carlos','Ana','Carlos','Noah', 'Ana'],
    'Vendas':[250,230,150,400,800]
}
```

```python
df_vendas = pd.DataFrame(dados_vendas)
```

```python
df_vendas
```

```python
# Criar um tabela dinÂmica para visualizar o total de vendas por região e vendedor
# No Python tudo é um obejto e todo objeto tem método(pivot-table) e parÂmetros (os valores em parenteses)
tabela_dinamica = df_vendas.pivot_table(values='Vendas',
                                        index='Região',
                                        columns='Vendedor',
                                        aggfunc='sum',
                                        fill_value=0)
```

```python
print('\n___________Tabela dinâmica: total de vendas por região e vendedor_____________\n')
tabela_dinamica
```
