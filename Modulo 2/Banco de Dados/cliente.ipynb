```python
import sqlite3 
```

```python
cliente = 'clientes.db'
```

# Criar tabelas

```python
script_clientes = '''CREATE TABLE IF NOT EXISTS Clientes(
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT NOT NULL,
                idade INTEGER NOT NULL,
                email TEXT NOT NULL
)'''

try:
    with sqlite3.connect(cliente) as con:

        cur = con.cursor()

        cur.execute(script_clientes)

        con.commit()
        print("Tabela criada com sucesso!!")
except sqlite3.OperationalError as e:
    print('Erro:', e)
    
```

# Consultar tabelas

```python
sql = 'SELECT * FROM Clientes'

try:
    with sqlite3.connect(cliente) as con:
        
        cur = con.cursor()
        
        cur.execute(sql)
        res = cur.fetchall()
        print(res)
except sqlite3.OperationalError as e:
    print ('Erro', e)
```

# Inserir Dados

```python
sql = 'INSERT INTO Clientes (nome, idade, email) VALUES (?,?,?)'

try:
    with sqlite3.connect(cliente) as con:

        cur = con.cursor()

        cur.execute(sql, ('Alexsandra', 20, 'alexsandra@gmail.com'))

        con.commit
except sqlite3.OperationalError as e:
    print('Erro:', e)
```

# Atualizar Registro

```python
sql = 'UPDATE Clientes SET idade = ? WHERE id = ?'

try:
    with sqlite3.connect(cliente) as con:

        cur= con.cursor()

        cur.execute(sql,(28, 1))

        con.commit()
except sqlite3.OperationalError as e:
    print('Erro', e)
```

# Deletar Registro

```python
sql = 'DELETE FROM Clientes WHERE id = ?'

try:
    with sqlite3.connect(cliente) as con:
        cur= con.cursor()
        cur.execute(sql,(2,))
        con.commit()
except sqlite3.OperationalError as e:
    print('Erro', e)
```
